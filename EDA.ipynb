```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
directory = 'archive'
data_dictionary = {}
# iterate over files in
# that directory
for filename in os.scandir(directory):
    if filename.is_file():
        data_dictionary[filename.name.split(".")[0]] = pd.read_csv(filename)
```

```python
df1 = data_dictionary['games']
df2 = data_dictionary['teams']
```

```python
df3 = df1[['SEASON','GAME_DATE_EST','HOME_TEAM_ID','VISITOR_TEAM_ID','PTS_home', 'FG_PCT_home','FT_PCT_home', 'FG3_PCT_home', 'AST_home', 'REB_home', 'TEAM_ID_away','PTS_away', 'FG_PCT_away', 'FT_PCT_away', 'FG3_PCT_away', 'AST_away','REB_away','HOME_TEAM_WINS']]
```

```python
df4 = df2[['TEAM_ID','ABBREVIATION',
       'NICKNAME']]
```

```python
def find_abv(identification):
    return df4[df4['TEAM_ID'] == identification].iloc[0]['ABBREVIATION']

def find_nickname(identification):
    return df4[df4['TEAM_ID'] == identification].iloc[0]['NICKNAME']
```

```python
df3['HOME_ABV'] = df3['HOME_TEAM_ID'].map(find_abv)
df3['HOME_NICKNAME'] = df3['HOME_TEAM_ID'].map(find_nickname)
df3['AWAY_ABV'] = df3['VISITOR_TEAM_ID'].map(find_abv)
df3['AWAY_NICKNAME'] = df3['VISITOR_TEAM_ID'].map(find_nickname)
```

```python
final_df = df3[['SEASON','PTS_home','FG_PCT_home', 'FT_PCT_home', 'FG3_PCT_home', 'AST_home', 
     'REB_home','PTS_away', 'FG_PCT_away', 'FT_PCT_away','FG3_PCT_away', 'AST_away', 'REB_away', 'HOME_TEAM_WINS', 
     'HOME_ABV','HOME_NICKNAME', 'AWAY_ABV', 'AWAY_NICKNAME']]
```
